```python
import pandas as pd
import numpy as np
f_ = 'data/UrbanSound8K/metadata/UrbanSound8Koriginal.csv'
meta = pd.read_csv(f_)
```

```python
#meta['split'] = np.random.randn(meta.shape[0], 1)

msk = np.random.rand(len(meta)) <= 0.8

train = meta[msk]
test = meta[~msk]
```

```python
test # test set
```

```python
# moving test files to audio folder
import shutil
root='data/UrbanSound8K/audio_22_5'
for i in range(len(test)):
    path_source=root+'/fold'+str(test.values[i,5])+'/'+str(test.values[i,0])
    path_dest=root+'/fold0/'+str(test.values[i,0])
    shutil.move(path_source,path_dest)
```

```python
# modifying the fold of test set
test.loc[:, 'fold'] = 0
```

```python
metaNEW = pd.concat([train,test])
metaNEW.to_csv('data/UrbanSound8K/metadata/UrbanSound8K.csv')
```
